# ***Importing packages***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## ***Reading the dataset***

```python
data = pd.read_csv('/content/globalterrorismdb_0718dist.csv', encoding='ISO-8859-1')
```

# ***Data Exploration***

```python
data.head()
```

```python
data.describe()
```

```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                     'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                     'nwound': 'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                     'weaptype1_txt':'Weapon_type','motive':'Motive'
                    },inplace=True)
```

```python
data = data[['Year','Month','Day','Country','Region','state','city','latitude','longitude','AttackType','Killed','Wounded','Target',
            'Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.info()
```

# ***Data Cleaning***

```python
data.dropna(axis=0,inplace = True)
```

```python
data.isnull().sum()
```

```python
data
```

# ***Data visualization***

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Year', data=data)
plt.xticks(rotation=45)
plt.title('Terrorist Attacks by Year')
plt.show()
```

```python
plt.figure(figsize=(30, 5))
sns.countplot(x='Country', data=data, order=data['Country'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.title('Most Frequent Countries in Terrorist Attacks')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Region', data=data, order=data['Region'].value_counts().index)
plt.xticks(rotation=45)
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.title('Most Frequent Regions in Terrorist Attacks')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='AttackType', data=data, order=data['AttackType'].value_counts().index)
plt.xticks(rotation=45)
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.title('Most Frequent Attack Types in Terrorist Attacks')
plt.show()
```
